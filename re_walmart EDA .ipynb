1. find different payment method and number of transactions, number of quantity sold

```python
payment_group = df.groupby('payment_method')

transaction_counts = payment_group['invoice_id'].count()
print("Number of Transactions per Payment Method:\n", transaction_counts)
```

```python
quantity_sold = payment_group['quantity'].sum()
print("\nTotal Quantity Sold per Payment Method:\n", quantity_sold)
```

```python
payment_methods = df['payment_method'].unique()
print("\nDifferent Payment Methods:\n", payment_methods)
```

```python
payment_summary = payment_group.agg(
    num_transactions=('invoice_id', 'count'),
    total_quantity=('quantity', 'sum')
)
print("\nPayment Summary:\n", payment_summary)
```

2. Identify the highest-rated category in each branch, displaying the branch, category

```python
highest_rated = df.groupby('Branch')['rating'].transform('max') == df['rating']
highest_rated_categories = df[highest_rated][['Branch', 'category']] 
highest_rated_categories
```

```python
highest_rated = df.groupby('Branch')['rating'].max().reset_index()

highest_rated_categories = pd.merge(highest_rated, df, on=['Branch', 'rating'], how='left')

result = highest_rated_categories[['Branch', 'category']]

result
```

3. Identify the business day for each branch based on the number of transaction

```python
daily_transactions = df.groupby(['Branch', 'date']).size().reset_index(name='transaction_count')
busiest_days = daily_transactions.loc[daily_transactions.groupby('Branch')['transaction_count'].idxmax()]
busiest_days
```

4.Determine the average, minimum, and maximum rating of products for each city, List the city,average_rating, and maxx_rating

```python
city_ratings = df.groupby('City')['rating'].agg(['mean', 'min', 'max']).reset_index()
city_ratings = city_ratings.rename(columns={'mean': 'average_rating', 'min': 'min_rating', 'max': 'max_rating'})
city_ratings
```

5 .calculate the total profit for each category by considering total_profit as unit_price * quantity * profit_margin, list category and total_profit, ordered from highest to lowest profit

```python
df['total_profit'] = df['unit_price'] * df['quantity'] * df['profit_margin']
category_profit = df.groupby('category')['total_profit'].sum().reset_index()
category_profit_sorted = category_profit.sort_values(by='total_profit', ascending=False)
category_profit_sorted
```

6. Determine the most common payment method for each Branch, Display Branch and the preferred payment method

```python
payment_counts = df.groupby(['Branch', 'profit_margin']).size().reset_index(name='transaction_count')
most_common_payments = payment_counts.loc[payment_counts.groupby('Branch')['transaction_count'].idxmax()]
print(most_common_payments[['Branch', 'profit_margin']])
```
